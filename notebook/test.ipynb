```python
import pickle
with open("vectorise.pkl","rb") as f:
    vectorise = pickle.load(f)
with open("model.pkl","rb") as s:
    model = pickle.load(s)
```

```python
def predict(text):
    transform = vectorise.transform([text])
    return "positive" if model.predict(transform) == 1 else "negative"
```

```python
predict("This movie was an emotional masterpiece. The performances felt real, and the story stayed with me long after it ended.")
```

```python
ls = [
    "This movie was an emotional masterpiece. The performances felt real, and the story stayed with me long after it ended.",
    "A beautifully shot film with great acting and an engaging plot. I enjoyed every minute of it",
    "Absolutely loved it! The pacing, music, and characters were all perfect. Would definitely recommend.",
    "Surprisingly good! I didn't expect much, but it turned out to be funny, heartwarming, and well-written.",
    "One of the best movies I've seen this year. The cinematography and soundtrack were stunning"
]
```

```python
for rev in ls:
    print(predict(rev))
```
